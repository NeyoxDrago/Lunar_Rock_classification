```
import os
import cv2
import numpy as np

# shapes={}
# for i in os.listdir("Train Images/Large/"):
#   shape = np.array(cv2.imread("Train Images/Large/"+i)).shape
#   if shape not in shapes:
#     shapes[shape]=0
#   else:
#     shapes[shape]+=1
```

```
train_labels=[]
train_data=[]
test_labels =[]
test_data=[]

classes = {"Large":0,"Small":1}

for folder in os.listdir("Train Images"):
  for file in os.listdir(os.path.join("Train Images",folder)):
    
    image = cv2.imread(os.path.join("Train Images",folder,file))
#     image = cv2.resize(image,(360,240))
    train_labels.append(classes[folder])
    train_data.append(image)
```

```
# import pickle
# with open("train_data.pickle","wb") as file:
#   pickle.dump(train_data,file)
```

```
# np.array(train_data).shape
train_data=np.array(train_data)
```

```
train_data.shape
```

```
from keras.utils import to_categorical

train_labels = to_categorical(train_labels,num_classes=2)
```

```
from keras.layers import Conv2D , MaxPooling2D , Input , Dense , Dropout , GlobalMaxPooling2D,Flatten,BatchNormalization
from keras.models import Model
```

```
input_ = Input(shape=(480,720,3,))

Conv =  Conv2D(8,(2,2),data_format="channels_last",activation="relu")(input_)
Conv2 = Conv2D(16,(2,2),data_format="channels_last",activation="relu")(Conv)
Conv2 = Conv2D(16,(2,2),data_format="channels_last",activation="relu")(Conv2)
output= MaxPooling2D(pool_size=(2,2))(Conv2)

output = BatchNormalization()(output)

Conv2 = Conv2D(16,(2,2),data_format="channels_last",activation="relu")(output)
Conv2 = Conv2D(32,(2,2),data_format="channels_last",activation="relu")(Conv2)
Conv2 = Conv2D(32,(2,2),data_format="channels_last",activation="relu")(Conv2)
output= MaxPooling2D(pool_size=(2,2))(Conv2)
output = BatchNormalization()(output)

Conv2 = Conv2D(32,(2,2),data_format="channels_last",activation="relu")(output)
Conv2 = Conv2D(64,(2,2),data_format="channels_last",activation="relu")(Conv2)
Conv2 = Conv2D(64,(2,2),data_format="channels_last",activation="relu")(Conv2)
output= MaxPooling2D(pool_size=(2,2))(Conv2)
output = BatchNormalization()(output)

Conv2 = Conv2D(64,(2,2),data_format="channels_last",activation="relu")(output)
Conv2 = Conv2D(128,(2,2),data_format="channels_last",activation="relu")(Conv2)
Conv2 = Conv2D(128,(2,2),data_format="channels_last",activation="relu")(Conv2)
output= MaxPooling2D(pool_size=(2,2))(Conv2)
output = BatchNormalization()(output)

Conv2 = Conv2D(128,(3,3),data_format="channels_last",activation="relu")(output)
Conv2 = Conv2D(128,(2,2),data_format="channels_last",activation="relu")(Conv2)
Conv2 = Conv2D(128,(2,2),data_format="channels_last",activation="relu")(Conv2)
output= MaxPooling2D(pool_size=(2,2))(Conv2)
output = BatchNormalization()(output)

output= Dropout(0.2)(output)

output = Flatten()(output)
output= Dense(512 , activation="relu")(output)
output= Dropout(0.2)(output)
output= Dense(128 , activation="relu")(output)
output= Dropout(0.2)(output)
output= Dense(32 , activation="relu")(output)
output= Dropout(0.5)(output)
output= Dense(2,activation="softmax")(output)
```

```
model = Model(input_,output)
model.summary()
```

```
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['acc'])
```

```
model.fit(train_data,train_labels,epochs=30,verbose=1,batch_size=32)
```

```
model.save("custom.hdf5")
```

```
from keras.applications.vgg19 import VGG19
from keras.applications.resnet import ResNet50
from keras.applications.mobilenet import MobileNet
from keras.applications.densenet import DenseNet121
from keras.applications.xception import Xception
from keras.applications.inception_v3 import InceptionV3
from keras.layers import GlobalAveragePooling2D,Flatten
```

```
# base_model = VGG19(include_top=False,weights='imagenet',input_shape=(480,720,3))
base_model = ResNet50(include_top=False,weights='imagenet',input_shape=(480,720,3))
# base_model = Xception(include_top=False,weights='imagenet',input_shape=(480,720,3))
# base_model = InceptionV3(include_top=False,weights='imagenet',input_shape=(240,360,3))
# base_model = MobileNet(include_top=False,weights='imagenet',input_shape=(480,720,3))
# base_model = DenseNet121(include_top=False,weights='imagenet',input_shape=(240,360,3))
x = base_model.output
```

```
# x = GlobalAveragePooling2D()(x)
x = Flatten()(x)
x = Dense(128,activation="relu")(x)
x = Dropout(0.2)(x)
x = Dense(32, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(2,activation="softmax")(x)
```

```
model = Model(base_model.input , x)
```

```
model.summary()
```

```
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['acc'])
```

```
model.fit(train_data,train_labels,epochs=5,batch_size=2,verbose=1)
```

```
def predict(image):
#   print(image)
  img = cv2.imread("Test Images/"+image)
#   img = cv2.resize(img,(360,240))
  img = np.array([img])
  prediction = model.predict(img)
  result = np.argmax(prediction[0],axis=0)
  return result
  
predict("lg 988 (1).png")
```
